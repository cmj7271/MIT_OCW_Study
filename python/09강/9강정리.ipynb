* 데이터의 은닉화 : 1. 원치 않는 동작을 방지하는 역할([방어적프로그래밍](https://ko.wikipedia.org/wiki/%EB%B0%A9%EC%96%B4%EC%A0%81_%ED%94%84%EB%A1%9C%EA%B7%B8%EB%9E%98%EB%B0%8D))  
보통은  
`public` : class 외부에서도 접근 가능  
`protected` : 외부는 불가능, 자신과 자식 class 에서 접근 가능  
`private` : 외부, 자식 class 불가능, 자기자신 class 만 가능  
이라는 키워드를 통해 구현하지만...  
python 은 이런 키워드가 없으며, default 가 `public` 이다.  
따라서 기본적으로 외부에서 접근 가능하지만, 데이터 은닉화 관점에서 다른 언어에서는 접근 불가능한것처럼 필요한 경우 getter 함수를 통해 얻는 것이 좋다.  
이러한 프로그래밍은 데이터에 대한 접근 방식을 한정해주기 때문에 유지, 보수를 좀 더 수월하게 해준다.  

ex) data attribute 로 self.age 을 사용 중  
근데 우연히 코딩 중 서로 다른 class 가 같은 data attribute 를 쓰고, 혼합해서 사용되는 경우

````python
class age_1 :
    def __init__(self, age) : # int형을 기대
        self.age = age

    def get_age(self) :
        return self.age

    def __add__(self, other) :  # int 형이어야만 함
        return age_1(self.age + other.age)

    def __str__(self) :
        return "age1 : " + str(self.age)


class age_2 :
    def __init__(self, age) : # str형(영문)을 기대
        self.age = age

    def get_age(self) :
        return self.age

    def __str__(self) :
        return "age2 : " + str(self.age)

#----------------

a = age_1(7)
a_2 = age_1(8)

a_3 = a + a_2

b = age_2("seven")

print(a)
print(a_2)
print(a_3)
print(b)
print()

a.age = "seven"
b.age = 7

print(a)
print(a_2)
print(b)

print(a + a_2)
````

* 상속  
상위 개념과 하위 개념의 관계  
다양한 class 간의 공통점을 뽑아 추상화시킨다.  
추상화 시킨 내용은 개별 class 에서 재정의(overriding) 할 수도, 상위 class의 method를 그대로 사용할 수도 있다.  
ex) 동물 - 강아지 - 시베리안 허스키   

````python
class character(object) :
    def __init__(self) :
        self.hp = 100 # 모두 공통인 부분
        self.atk = 0  # 각 캐릭터마다 다르므로 재정의 필요
        self.name = ""
        self.guard = -1 
        # -1 : 가드 x, 0 : 하단, 1 : 중단, 2 : 상단
        self.grab_dmg = 10
        self.skill_1_dmg = 0

    # data attribute setter, getter 설정
    def set_hp(self, hp) :
        self.hp = hp
    def get_hp(self) :
        return self.hp
    
    def set_atk(self, atk) :
        self.atk = atk
    def get_atk(self) :
        return self.atk

    def set_name(self, name) :
        self.name = name
    def get_name(self) :
        return self.name

    def set_guard(self, num) :
        self.guard = num
    def get_guard(self) :
        return self.guard

    def set_grab_dmg(self, dmg) :
        self.grab_dmg = dmg
    def get_grab_dmg(self) :
        return self.grab_dmg

    def set_skill_1_dmg(self, dmg) :
        self.skill_1_dmg = dmg
    def get_skill_1_dmg(self) :
        return self.skill_1_dmg

    def attack(self, other) :
        if other.get_guard() != 0 : # 기본공격은 중단
            other.set_hp(other.get_hp() - self.get_atk())
            return self.get_atk()
        else :
            return -1

    def skill_1(self, other) : # 캐릭터마다 다르므로 캐릭터에서 따로 정의 필요
        return -2 # skill 없음

    def grab(self, other) :
        other.set_hp(other.get_hp() - self.get_grab_dmg()) # 잡기 기술은 공통적으로 데미지가 같음
        return self.get_grab_dmg()

    def atk_interface(self, other) :
        print(f"-1 : guard   0 : grab({self.get_grab_dmg()} damage)    1 : attack({self.get_atk()} damage)    2 : skill 1({self.get_skill_1_dmg()} damage)")
        commend = int(input())
        
        if commend == -1 :
            dmg = 0
            print("0 : low, 1 : mid, 2 : high")
            guard = int(input())
            self.set_guard(guard)
        
        elif commend == 0 :
            dmg = self.grab(other)
            print(str(self.get_name()) + " grabed " + str(other.get_name()) + " with " + str(self.get_grab_dmg()) + "damage.")
                
        elif commend == 1 :
            dmg = self.attack(other)
            if dmg != -1 :
                print(str(self.get_name()) + " attacked " + str(other.get_name()) + " with " + str(self.get_atk()) + " damage.")
        
        elif commend == 2 :
            dmg = self.skill_1(other)
            if dmg != -1 :
                print(f"{self.get_name()} use skill_1 to {other.get_name()} and got {dmg} damage.")
                
        else :
            dmg = 0
            print("commend miss!")

        if(commend > -1) : self.set_guard(-1) # 공격후에는 가드가 풀림
        if(dmg == -1) : print("attack failed (because of guard)")
        elif(dmg == -2) : print(f"attack failed (there is no skill to {self.get_name()})")

    def is_defeat(self) :
        return (self.get_hp() <= 0) 

    def __str__(self) :
        return str(self.get_name()) + " remained hp : " + str(self.get_hp())


class kazuya(character) :
    def __init__(self) :
        character.__init__(self)
        self.set_atk(40)
        self.set_name("kazuya")
        self.set_skill_1_dmg(50)
        self.transform = True # kazuya는 다른 캐릭터와는 달리 변신가능

    def skill_1(self, other) :
        if other.get_guard() != 2 :
        # 상단 공격임
            dmg = self.get_skill_1_dmg()
            if(self.transform) :
                dmg += 10
                self.transform = False
                print("kazuya is transformed")
            other.set_hp(other.get_hp() - dmg) # 좀 더 아픔 아야
            return dmg
        else :
            return -1


class king(character) :
    def __init__(self) :
        character.__init__(self)
        self.set_atk(30)
        self.set_name("king")
        self.set_grab_dmg(30)
        self.set_skill_1_dmg(40)

    def skill_1(self, other) :
        if other.get_guard() != 0 : # 하단공격임
            dmg = self.get_skill_1_dmg()
            other.set_hp(other.get_hp() - dmg)
            return dmg
        else : 
            return -1

player_1 = kazuya()
player_2 = king()

atk_seq = 1

while not player_1.is_defeat() and not player_2.is_defeat() :
    if atk_seq == 1 :
        print("player_1 turn\n")
        player_1.atk_interface(player_2)
        atk_seq = 2
    else :
        print("player_2 turn\n")
        player_2.atk_interface(player_1)
        atk_seq = 1

    print(player_1)
    print(player_2)
    print()

if(player_1.is_defeat()) : 
    print(f"player_1 {player_1.get_name} defeat")
else :
    print(f"player_2 {player_2.get_name()} defeat")
````

사람은 이 기계와 버튼으로만 소통할 수 있다.  
--> 즉 내부적으로 어떻게 돌아가는지 모름  
--> 메세지를 받은 객체가 어떻게 행동하지는 객체만의 자율성임  
기계는 다른 기계와 링크되어 소통하여 함께 문제를 해결할 수 있다.  
앨리스라는 기계와 음류수라는 기계가 있을 때,  
앨리스기계에서 음류수를 마신다라는 버튼을 누르면,  
사용자는 모르지만(즉, 기계에게 자율성을 맡긴다),  
음류수를 마신다는 행위는 앨리스라는 객체의 상태(예컨데 목마름, 키 등등) 을 변경시키고, 그 과정에서 음류수에게 마셔진다 라는 메세지를 전달하게 된다.  
역시 이 과정에서 음류수가 내부적으로 어떻게 처리하는지는 알 수 없다.

특정 예시에서 `일반화`시키는 느낌..?  
반복되는 특정 행동 --> 함수

but... 내가 읽는 책에서...  
추상화 == 현실세계에 대한 모방 ==> x

이지만, 추상화를 단순히 모방이라고 부르기에는 문제가 있다.  
예시로, 소프트웨어에서의 상품은 스스로 가격 계산이라는 행동을 취할 수 있다.  
이러한 점에서 필자(객체지향의 사실과 오해의)는 레베카 워프스브록의 `의인화`라는 표현을 제시한다.  
현실세계를 뛰어넘는 더 전지전능한 객체가 사용될 수 있다.  
그렇다면 현실을 참조해서 객체를 만들라는 것은 어떤 이유인가?  
현실과 소프트웨어간의 관계를 `은유`로 보기위해서이다.  
비록 음류수가 스스로 마셔질 수는 없지만, 음류수라는 이름을 통해, 즉 비유를 통해 그 상태나 행동을 유추 및 기억하기 쉽다는 의미이다.  
즉, 현실을 이용할 수 있다면 이용하되, 거기에 구속되지 말아라...라는 뜻인 듯하다...

* 다형성  
서로 다른 타입에 속하는 `객체`들이 동일한 `메세지`에 서로 다른 `메서드`를 이용해 `메세지`를 처리하는 것  
메세지?  
송신자가 수신자에게 `행동`(`책임`)을 요구하는 것  
여기에 수신자가 반응할 수 있다는 것은 `메세지`를 처리할 `책임`이 있다는 이야기  
이 때 송신자는 메세지의 수신자가 어떤이인지 신경쓰지 않는다. 다만 메세지의 내용을 처리할 수만 있으면 된다.  
ex) 왕이 모자장수, 엘리스, 토끼든간에 `증언하라`는 메세지를 전달한다는 것은 `증언`만 할 수 있으면 그만이라는 것이다.  
그것이 누가하는지는 왕이 신경쓰지 않는다.  
이렇게 전달받은 메세지에 따라 수신자는 자신의 `메서드`에 따라 메세지의 내용을 이행한다.  
송신자(왕)가 `증인` 역할을 할 수 있는 수신자이기만 하면 신경쓰지 않는다.  
이는 증인이라는 개념에 속하기만 하면 되므로 어떤 관점에서는 `캡슐화`라고 볼 수 있다.  
증인이라는 `개념`으로 묶여있기만 하고, 정확히 무엇인지는 상관없기 때문이다.  
개념 : 객체들을 특정 정의로 묶어놓은 것  
개념은 심볼, 외연, 내연으로 이루어짐  
심볼 --> 이름  
외연 --> 예시들    
내연 --> 정의  
위의 코드를 예시로 들면,  
`skill_1` 은 `character` 이기만 한다면 상관없음  
`player_1` 이 `player_2` 에게 `skill_1`(메세지) 를 전달, 혹은 `player_2` 가 `player_1`에게  
이는 어느 쪽이든 skill_1 이라는 `메세지`를 `책임`을 갖고 해결하면 된다.  
실제로 `player_1`, `player_2` 의 내부 구조, 즉 처리방식은 다르다.
