```c
#include <stdio.h>
#include <stdlib.h> // malloc() 을 사용하기 위한

#define MAX_ARR_SIZE 1048575

int rc(int** arr, int n);


int main(void) {
    // 최대 경우의 수가 너무 크기 때문에 이중배열 동적 할당을 해줘야 한다.
    int** arr = NULL;
    arr = (int**)malloc(sizeof(int*) * MAX_ARR_SIZE); // 2^20 - 1 = 1048575
    arr[0] = (int*)malloc(sizeof(int) * MAX_ARR_SIZE * 2);
    for(int i = 1; i < MAX_ARR_SIZE; i++) arr[i] = arr[i - 1] + 2;
    
    arr[0][0] = 1;
    arr[0][1] = 3;

    int num = 0;
    scanf("%d", &num);
    int length = rc(arr, num);

    printf("%d\n", length);

    for(int i = 0; i < length; i++) {
        printf("%d %d\n", arr[i][0], arr[i][1]);
    }
    // 동적할당 해체
    free(arr[0]);
    free(arr);

    return 0;
}


int rc(int** arr, int n) {
    if(n == 1) {
        return 1;
    }
    
    int length = rc(arr, n - 1);

    for(int i = 0; i < length; i ++) {
        for(int j = 0; j < 2; j ++) {
            if(arr[i][j] == 1) arr[i + length + 1][j] = 2;
            else if(arr[i][j] == 2) {
                arr[i][j] = 3;
                arr[i + length + 1][j] = 1;
            }
            else if(arr[i][j] == 3) {
                arr[i][j] = 2;
                arr[i + length + 1][j] = 3;
            }
        }
    }

    arr[length][0] = 1;
    arr[length][1] = 3;

    return 2*length + 1;
}
```  
c 언어 - 1

```c
#include <stdio.h>
#include <math.h>


void hanoi(int fir, int sec, int thi,  int n);

int main(void) {
    int num = 0;
    scanf("%d", &num);
    printf("%d\n", (int) pow(2, num) - 1);
    hanoi(1, 2, 3, num);


    return 0;
}

void hanoi(int fir, int sec, int thi, int n) {
    if(n == 1) {
        printf("%d %d\n", fir, thi);
        return;
    }

    hanoi(fir, thi, sec, n - 1);
    printf("%d %d\n", fir, thi);
    hanoi(sec, fir, thi, n - 1);
    
    return;
}
```

* 2630 색종이 만들기

```c
#include <stdio.h>
#include <stdlib.h>

void f(int start_row, int start_col, int length, int** arr, int* p_white, int* p_blue);

int main(void) {
    int n = 0;
    scanf("%d", &n);

    int white = 0, blue = 0;
    int* p_white = &white;
    int* p_blue = &blue;

    int** arr = (int**) malloc(sizeof(int*) * n);
    arr[0] = (int*) malloc(sizeof(int) * n*n);
    for(int i = 1; i < n; i++) {
        arr[i] = arr[i - 1] + n;
    }
    
    for(int i = 0; i < n; i++) {
        for(int j = 0; j < n; j++) {
            scanf("%d", &arr[i][j]);
        }
    }

    f(0, 0, n, arr, p_white, p_blue);

    printf("%d\n", white);
    printf("%d", blue);

    free(arr[0]);
    free(arr);
    return 0;
}


void f(int start_row, int start_col, int length, int** arr, int* p_white, int* p_blue) {
    int indi = arr[start_row][start_col];

    printf("\n%d %d %d %d\n", start_row, start_col, indi, length);

    for(int row = start_row; row < start_row + length; row++) {
        for(int col = start_col; col < start_col + length; col++) {
            if(arr[row][col] != indi) {
                int new_row = start_row + length/2;
                int new_col = start_col + length/2;

                f(start_row, start_col, length/2, arr, p_white, p_blue);
                f(new_row, start_col, length/2, arr, p_white, p_blue);
                f(start_row, new_col, length/2, arr, p_white, p_blue);
                f(new_row, new_col, length/2, arr, p_white, p_blue);

                return;
            }
        }
    }

    if(indi == 0) {
        *p_white += 1;
        return;
    }
    *p_blue += 1;
    return;
}
```
